```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('mixamo_data.npz', allow_pickle=True)

bone_heads = data['bone_heads']
bone_parents = data['bone_parents']
bone_names = data['bone_names']

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# plot joints
ax.scatter(
    bone_heads[:, 0],
    bone_heads[:, 1],
    bone_heads[:, 2],
    s=20
)

# draw bone connections
for i, parent in enumerate(bone_parents):
    if parent != -1:
        x = [bone_heads[i, 0], bone_heads[parent, 0]]
        y = [bone_heads[i, 1], bone_heads[parent, 1]]
        z = [bone_heads[i, 2], bone_heads[parent, 2]]
        ax.plot(x, y, z)

# optional: label some bones
for i, name in enumerate(bone_names[:15]):
    ax.text(
        bone_heads[i, 0],
        bone_heads[i, 1],
        bone_heads[i, 2],
        str(name),
        fontsize=8
    )

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Skeleton / Bone Heads")
plt.show()
```
